```python
import datetime as dt
```

```python


def conv_date(date):
    formats = ["%Y", "%d %b %Y", "%b %d, %Y","%b %Y"]
    for fmt in formats:
        try:
            return dt.datetime.strptime(date, fmt)
        except ValueError:
            pass
    raise ValueError(f"time data '{date}' does not match any of the formats {formats}")
```

```python
pop_data["Date"]=pop_data["Date"].apply(conv_date)
```
